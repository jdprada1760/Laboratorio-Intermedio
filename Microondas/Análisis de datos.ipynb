```python
%pylab inline
from scipy import optimize
!ls
```

```python
# Carga los datos de la curva de resonancia
name = "polarizador.data"
table = np.loadtxt(name, delimiter = ',').T
```

```python
# Grafica
scatter(table[0], table[1])
```

```python
def fun(x,A):
    x = (x)*np.pi/180    
    return A*((cos(x))**4)
params, errors = optimize.curve_fit(fun,table[0] ,table[1], maxfev = 10000)
print params, errors
```

```python
# Grafica el fit contra los datos 
x = linspace(0,90, 200)
y = fun(x, table[1][0])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
ax.errorbar(table[0], table[1], fmt = "o",yerr = 0.2, xerr = 5, markersize = 7   )
ax.plot(x,y)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.set_ylabel("$Intensidad (10^{-1}mA)$", fontsize = 27)
ax.set_xlabel("$Angulo (^o)$", fontsize = 27)
ax.set_title("$Curva\ de\ polarizacion$", fontsize = 27)
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
                # specify integer or one of preset strings, e.g.
                #tick.label.set_fontsize('x-small') 
fig.savefig('polarizador2.jpg')
```

```python
#d = 10
#a = 1.5
#l = 2.857
A = 10
def inter(x,l, d):
    x = x*np.pi/180
    y = np.pi*d*sin(x)/l
    rta=  A*(sin(y)/y+0.001)**2
    return rta
```

```python
params, errors = optimize.curve_fit(inter,table[0] ,table[1], p0 = [2.8,1.5], maxfev = 1000000)
print params, errors
```

```python
# Grafica el fit contra los datos 
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
ax.errorbar(table[0], table[1], fmt = "o",yerr = 0.2, markersize = 9  )
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.set_ylabel("$Intensidad (10^{-1}mA)$", fontsize = 27)
ax.set_xlabel("$Angulo (^o)$", fontsize = 27)
ax.set_title("$Patron\ de\ interferencia\ d = 10.5cm$", fontsize = 27)
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
                # specify integer or one of preset strings, e.g.
                #tick.label.set_fontsize('x-small') 
fig.savefig('inter1.jpg')
```

```python
min1 = np.sin(np.pi*np.array([10,30,45])/180)
min2 = np.sin(np.pi*np.array([15,35,60])/180)
```

```python
a = np.array([1,2,3])
```

```python
scatter(a,min1)
scatter(a,min2)
```

```python
def lin(x,m,b):
    return m*x+b
params1, errors1 = optimize.curve_fit(lin, a, min1, p0 = [0,0.], maxfev = 1000000)
params2, errors2 = optimize.curve_fit(lin, a, min2, p0 = [0,0.], maxfev = 1000000)
print params1,params2
```

```python
# Grafica el fit contra los datos 
x = linspace(0,4,50)
y1 = lin(x,params1[0],params1[1])
y2 = lin(x,params2[0],params2[1])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
ax.errorbar(a, min1, fmt = "o",yerr = sqrt(1-min1**2)*3*np.pi/180, markersize = 9 , c = 'y',
            label = r"d1: $\sin{\theta} = 0.2667n - 0.073$" )
ax.errorbar(a, min2, fmt = "o",yerr = sqrt(1-min2**2)*3*np.pi/180, markersize = 9 , c = 'black',
            label = r"d2: $\sin{\theta} = 0.30036n - 0.041$")
plot(x,y1)
plot(x,y2)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.set_ylabel(r"$\sin{ \theta }$", fontsize = 27)
ax.set_xlabel("$n$", fontsize = 27)
ax.set_title("$Relacion\ de\ interferencia$", fontsize = 27)
lgd = ax.legend(loc='lower center',bbox_to_anchor=(0.75, 0.1),fancybox=True, shadow=True, fontsize = 25)
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
                # specify integer or one of preset strings, e.g.
                #tick.label.set_fontsize('x-small') 
fig.savefig('inter3.jpg', bbox_extra_artist=[lgd])
```
