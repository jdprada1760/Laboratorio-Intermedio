```python
%pylab inline
from scipy import optimize
```

# Calibracion

```python
#Posiciones de las lineas del helio
x = array([14.0,12.1,8.6,5.2])
#Longitudes de onda teoricas de cada linea
l = array([ 667.815, 587.562, 501.567, 447.18 ])
# Grafica para ver verificar la relacion
scatter(x,l)
```

# Claramente la relación no es lineal, un mejor fit sería con un polinomio de grado 2

```python
# Funcion lineal para calibrar lambda
def lambdalin(x,a,b):
    return a*x+b
# Funcion cuadratica para calibrar lambda
def lambdaquad(x,a,b,c):
    return a*x**2 + b*x + c
def lambdaquad2(x,a,c):
    return a*x**2  + c
# Gets the parameters 
params1, errors1 = optimize.curve_fit(lambdalin,x,l, sigma = 0.2, maxfev = 100000)
params2 , errors2 = optimize.curve_fit(lambdaquad,x ,l, sigma = 0.2, maxfev = 100000)
params3 , errors3 = optimize.curve_fit(lambdaquad2,x ,l, sigma = 0.2, maxfev = 100000)
print params2, [sqrt(errors2[0][0]),sqrt(errors2[1][1]),sqrt(errors2[2][2])]
print params3, [sqrt(errors3[0][0]),sqrt(errors3[1][1])]
```

```python
# Grafica ambos fits para ver cual es el mejor
y = linspace(4,16,100)
f1 = lambdalin(y, params1[0], params1[1])
f2 = lambdaquad(y, params2[0], params2[1], params2[2])
f3 = lambdaquad2(y, params3[0], params3[1])
#scatter(x,l)
# Grafica 1/lambda  contra 1/n**2
# Grafica el fit contra los datos 
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
ax.errorbar(x, l, fmt = "o",xerr  = 0.2 , markersize = 7 )
ax.plot(y,f1, label = r"$y = 1.88x^2 -11.66x + 459.06$")
ax.plot(y,f2, label = r"$y = 1.288x + 408.22$")
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
ax.set_ylabel(r"$\lambda(nm)$", fontsize = 30)
ax.set_xlabel(r"$d$", fontsize = 30)
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
        
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax.set_title(r"$Curva\ de\ calibraci\'on$", fontsize = 30)
lgd = ax.legend(loc='lower center',bbox_to_anchor=(0.8, 0.85),fancybox=True, shadow=True, fontsize = 17)
fig.savefig('calibre.jpg', bbox_extra_artist=[lgd])
#plot(y,f3, c = 'g')
```

```python
a = params2[0]
b = params2[1]
c = params2[2]
da = sqrt(errors3[0][0])
#db = sqrt(errors2[1][1])
dc = sqrt(errors3[1][1])
# Calcula el error asociado a cada medida teniendo en cuenta la propagacion de error y las incertidumbres del fit
def err(x,dx):
    ans = (da*x**2)**2  + dc**2 + (dx*(2*a*x))**2
    return sqrt(ans)
```

# Espectro del hidrogeno

```python
# Distancias medidas para el espectro del hidrogeno
hx = array([3.8,7.4,13.5])
# Longitudes de onda de las lineas del hidrogeno
hl = lambdaquad(hx,a,b,c)
# errores
dhl = err(hx,0.1)
print hl,dhl
hlteo = array([434.0572, 486.135, 656.279])
print (hl-hlteo)/hlteo
```

# Rydberg Constant

```python
# Numero asociado a la serie de balmer para cada hl
n = array([5.0,4.0,3.0])
y = 1/n**2
x = 1/hl
dx = dhl/(hl**2)
print x,dx
# Define la funcion de las series de balmer
def balmer(f,R):
    # k = 1/R
    return R*(1.0/4.0 - f)
# Realiza el fit
params, errors = optimize.curve_fit(balmer,y,x, sigma = dx, maxfev = 10000000)
print params, sqrt(errors), errors
```

```python
# Grafica 1/lambda  contra 1/n**2
# Grafica el fit contra los datos 
yy = linspace(0.03,0.12, 100)
xx = balmer(yy, params[0])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
ax.errorbar(x, y, fmt = "o",xerr  = dx , markersize = 7   )
ax.plot(xx,yy, label = r"$x = $" + "0.011038" + r"$\left(\frac{1}{4} -y\right)$")
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.set_ylabel(r"$\frac{1}{n^2} (10^{-1})$", fontsize = 30)
ax.set_xlabel(r"$\frac{1}{\lambda_H} (10^{-3}/nm)$", fontsize = 30)
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(20) 
        
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax.set_title(r"$Serie\ de\ Balmer$", fontsize = 30)
lgd = ax.legend(loc='lower center',bbox_to_anchor=(0.73, 0.82),fancybox=True, shadow=True, fontsize = 25)
fig.savefig('balmer.jpg', bbox_extra_artist=[lgd])
```
